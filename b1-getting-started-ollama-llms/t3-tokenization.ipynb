# Understaing the tokenization with code

## Require imports
* tiktoken is a fast [BPE](https://en.wikipedia.org/wiki/Byte_pair_encoding) tokeniser for use with OpenAI's models.

```python
import tiktoken
```

## Tokenize a given input
### Tiktoken:
* To get the tokeniser corresponding to a specific model in the **OpenAI API**.

### Refence
* https://github.com/openai/tiktoken

```python
enconding = tiktoken.encoding_for_model("gpt-4.1-mini")

input_text = "Hello, how are you doing today?"
tokens = enconding.encode(input_text)

print(tokens)
```

# Print tokens for each word

```python
for token_id in tokens:
    token_text = enconding.decode([token_id])
    print(f"Token ID: {token_id}, Token Text: '{token_text}'")
```
