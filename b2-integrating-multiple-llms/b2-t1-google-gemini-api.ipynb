# Google API to access LLMs

# Imports

```python
import os
from dotenv import load_dotenv
from google import genai
```

# Config setup

```python
load_dotenv()

api_key = os.getenv('GEMINI_API_KEY')
base_url = os.getenv('GEMINI_BASE_URL')
model = os.getenv('GEMINI_FM')

if not api_key and not base_url:
    print('API key or Base URL missing.')
else:
    print(f'Config loaded successfully. \nAPI Key: {api_key[:10]}.... \nBase URL: {base_url}')

```

# Connect with gemini

```python
client = genai.Client()

payload = "Describe the color Blue to someone who has never been able to see in 1 sentence"

resp = client.models.generate_content(model=model, contents=payload)

print(resp.text)
```
